```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the cleaned data after preprocessing.
df = pd.read_csv("C:/Users/KRISHNA/Desktop/SHWETA/Fall 2024/Project/Cleaned_Loan_approval_dataset.csv")
```

```python
# Combining the income from all the sources.
df['TotalIncome'] = (df['IncomeFromPrincipalEmployer'] + df['IncomeFromPension'] + 
                     df['IncomeFromFamilyAllowance'] + df['IncomeFromSocialWelfare'] + 
                     df['IncomeFromLeavePay'] + df['IncomeFromChildSupport'] + 
                     df['IncomeOther']+df['IncomeTotal'])
```

```python
df.drop(['IncomeFromPrincipalEmployer', 'IncomeFromPension',
       'IncomeFromFamilyAllowance', 'IncomeFromSocialWelfare',
       'IncomeFromLeavePay', 'IncomeFromChildSupport', 'IncomeOther',
       'IncomeTotal'],axis=1,inplace=True)
```

```python
# Summing up the liabilities.
df['TotalLiabilities'] = df['LiabilitiesTotal'] + df['ExistingLiabilities'] + df['RefinanceLiabilities']
```

```python
df.drop([ 'ExistingLiabilities', 'LiabilitiesTotal',
       'RefinanceLiabilities'],axis=1,inplace=True)
```

```python
df['TotalLiabilities'].describe()
```

```python
# Checking skewness of all the columns.
df1 = df.select_dtypes(include=['number'])
skewed_features = df1.apply(lambda x: x.skew()).sort_values(ascending=False)

# Plotting the skewness values
plt.figure(figsize=(25, 6))
sns.barplot(x=skewed_features.index, y=skewed_features.values)
plt.xticks(rotation=90) 
plt.title('Skewness of Each Feature')
plt.ylabel('Skewness')
plt.xlabel('Features')
plt.show()
```

```python
threshold = 2

# Selecting features to transform based on skewness
features_to_transform = skewed_features[skewed_features.abs() > threshold].index.tolist()

# Apply transformations
for column in features_to_transform:
    if skewed_features[column] > 0:
        if (df[column] >= 0).all(): 
            df[column] = np.log1p(df[column])
        else:
            min_value = df[column].min()
            df[column] = df[column] + abs(min_value) + 1 
            df[column] = np.log1p(df[column])
```

```python
# Checking skewness of all the columns.
df1 = df.select_dtypes(include=['number'])
skewed_features = df1.apply(lambda x: x.skew()).sort_values(ascending=False)

# Plotting the skewness values
plt.figure(figsize=(25, 6))
sns.barplot(x=skewed_features.index, y=skewed_features.values)
plt.xticks(rotation=90) 
plt.title('Skewness of Each Feature')
plt.ylabel('Skewness')
plt.xlabel('Features')
plt.show()
```

```python
def categorize_liabilities(value):
    if value <= 3.948162:
        return 'Low'
    elif value <= 6.329605:
        return 'Medium'
    else:
        return 'High'

df['LiabilityGroup'] = df['TotalLiabilities'].apply(categorize_liabilities)
```

```python
plt.figure(figsize=(10, 5))

# Boxplot for Age vs Default Probability
sns.boxplot(x='LiabilityGroup', y='ProbabilityOfDefault', data=df, palette='pastel')
plt.title('Default Probability by Liabilities')
plt.xlabel('Liabilties')
plt.ylabel('Default Probability')
plt.show()
```

```python
features = ["EmploymentDurationCurrentEmployer","Rating","CreditScoreEsMicroL","Restructured","Status","NewCreditCustomer"]
for feature in features:
    df[feature].value_counts().plot(kind='pie', figsize=(12,6), autopct="%1.2f%%")
    plt.title(feature)
    plt.show()
```

```python
def categorize_age(age):
    if age < 25:
        return '18-24'
    elif 25 <= age < 36:
        return '25-35'
    elif 36 <= age < 46:
        return '36-45'
    elif 46 <= age < 56:
        return '46-55'
    else:
        return '56+'

df['AgeGroup'] = df['Age'].apply(categorize_age)
```

```python
df['DefaultRisk'] = np.where(df['ProbabilityOfDefault'] >= 0.2, 'High', 'Low')

# Using FacetGrid to create a combined chart
g = sns.FacetGrid(df, col='DefaultRisk', height=5, aspect=1)
g.map_dataframe(sns.countplot, x='AgeGroup', hue='Gender', palette='pastel', alpha=0.7)
g.set_axis_labels('Age Group', 'Count of Borrowers')
g.set_titles(col_template='{col_name} Default Risk')
g.add_legend(title='Gender', loc='upper right')
plt.show()
```

```python
def categorize_expectedreturn(value):
    if value < 0:
        return 'Negative'
    elif value < 0.100773 :
        return 'Low'
    elif value <= 0.150562:
        return 'Moderate'
    else:
        return 'High'


df['ReturnGroup'] = df['ExpectedReturn'].apply(categorize_expectedreturn)
```

```python
plt.figure(figsize=(10, 5))

# Boxplot for Age vs Default Probability
sns.boxplot(x='ReturnGroup', y='ProbabilityOfDefault', data=df, palette='pastel')
plt.title('Default Probability by Expected Return')
plt.xlabel('Expected Return')
plt.ylabel('Default Probability')
plt.show()
```
