```python
def parse_input(request,encoder): # request is the json object.
    
    #data = json.loads(request)
     
     # Convert JSON data to DataFrame
    df = pd.DataFrame([request])
    
    categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
    for column in categorical_columns:
        df[column] = encoder.fit_transform(df[column])
     
     # check for skewness 
    skewed_features = df.apply(lambda x: x.skew()).sort_values(ascending=False)
    threshold = 2
    features_to_transform = skewed_features[skewed_features.abs() > threshold].index.tolist()
     
    # Apply transformations
    for column in features_to_transform:
        if skewed_features[column] > 0:
            if (df[column] >= 0).all(): 
                    df[column] = np.log1p(df[column])
            else:
                    min_value = df[column].min()
                    df[column] = df[column] + abs(min_value) + 1 
                    df[column] = np.log1p(df[column])
                    
    # Call the predict loan function
    loan_amount = predict_loan(df,loan_scaler_x,loan_scaler_y)
    df['Amount'] = loan_amount
    
    # Call the predict default probability function
    default_prob = predict_prob(df,defaultprob_scaler_x,defaultprob_scaler_y)
    df['ProbabilityOfDefault'] = default_prob
    
    # Call the predict expected return function
    exp_return = predict_expreturn(df,expreturn_scaler_x,expreturn_scaler_y)
    df['ExpectedReturn'] = exp_return
    
    # Call the predict interest function
    interest = predict_interest(df,interest_scaler_x,interest_scaler_y)
    df['Interest'] = interest
    
    # Insert into database
    # insert(df)
        # Write the fucntion for inserting this df into database......
    return df
```

```python
def predict_loan(df,loan_scaler_x,loan_scaler_y):
    
    # Considering the features for loan amount prediction.
    loan_features = ['NewCreditCustomer', 'VerificationType', 'Age', 'Gender',
       'AppliedAmount', 'UseOfLoan', 'EmploymentStatus',
       'EmploymentDurationCurrentEmployer', 'OccupationArea',
       'HomeOwnershipType', 'TotalIncome', 'TotalLiabilities', 'DebtToIncome',
       'FreeCash', 'Rating', 'CreditScoreEsMicroL', 'CreditScoreEeMini',
       'NoOfPreviousLoansBeforeLoan', 'AmountOfPreviousLoansBeforeLoan']
    
    loan_df = df[loan_features]
    X = pd.DataFrame(loan_scaler_x.fit_transform(loan_df), columns=loan_df.columns)
    X = X.values.reshape(1, -1)
    
    # Perform the prediction
    loan_model =  joblib.load('C:/Users/KRISHNA/Desktop/SHWETA/Fall 2024/Project/Models/loan_model.joblib')
    y_pred = loan_model.predict(X).reshape(-1, 1)
    
    loan_amount = loan_scaler_y.inverse_transform(y_pred)
    
    print("Loan Amount: "+ str(loan_amount))
    return loan_amount


def predict_prob(df,defaultprob_scaler_x,defaultprob_scaler_y):
    
    prob_features = ['AppliedAmount', 'Amount',
    'UseOfLoan', 'TotalLiabilities','TotalIncome',
    'DebtToIncome', 'FreeCash',
    'NewCreditCustomer', 'CreditScoreEsMicroL', 'CreditScoreEeMini',
    'NoOfPreviousLoansBeforeLoan', 'AmountOfPreviousLoansBeforeLoan',
    'Age', 'EmploymentStatus', 'EmploymentDurationCurrentEmployer', 'HomeOwnershipType',
    'Rating', 'Restructured']
    
    prob_df = df[prob_features]
    
    X = pd.DataFrame(defaultprob_scaler_x.fit_transform(prob_df), columns=prob_df.columns)
    X = X.values.reshape(1, -1)
    
    # Perform the prediction
    defaultprob_model =  joblib.load('C:/Users/KRISHNA/Desktop/SHWETA/Fall 2024/Project/Models/defaultprob_model.joblib')
    y_pred = defaultprob_model.predict(X).reshape(-1,1)
    
    default_prob = defaultprob_scaler_y.inverse_transform(y_pred)
    
    return default_prob

def predict_expreturn(df,expreturn_scaler_x,expreturn_scaler_y):
    
    expreturn_features = ['AppliedAmount', 'Amount','ProbabilityOfDefault',
    'UseOfLoan', 'TotalLiabilities','TotalIncome',
    'DebtToIncome', 'FreeCash',
    'NewCreditCustomer', 'CreditScoreEsMicroL', 'CreditScoreEeMini',
    'NoOfPreviousLoansBeforeLoan', 'AmountOfPreviousLoansBeforeLoan',
    'Age', 'EmploymentStatus', 'EmploymentDurationCurrentEmployer', 'HomeOwnershipType',
    'Rating', 'Restructured']
    
    expreturn_df = df[expreturn_features]
    
    X = pd.DataFrame(expreturn_scaler_x.fit_transform(expreturn_df), columns=expreturn_df.columns)
    X = X.values.reshape(1, -1)
    
    # Perform the prediction
    expreturn_model =  joblib.load('C:/Users/KRISHNA/Desktop/SHWETA/Fall 2024/Project/Models/expreturn_model.joblib')
    y_pred = expreturn_model.predict(X).reshape(-1,1)
    
    exp_return = expreturn_scaler_y.inverse_transform(y_pred)
    
    return np.log1p(exp_return)

def predict_interest(df,interest_scaler_x,interest_scaler_y):
    
    interest_features = ['AppliedAmount', 'Amount','UseOfLoan',
    'NewCreditCustomer', 'Age', 'Gender', 'EmploymentStatus',
    'EmploymentDurationCurrentEmployer', 'OccupationArea', 'TotalIncome',
    'TotalLiabilities', 'DebtToIncome', 'FreeCash',
    'HomeOwnershipType', 'CreditScoreEsMicroL', 'CreditScoreEeMini',
    'NoOfPreviousLoansBeforeLoan', 'AmountOfPreviousLoansBeforeLoan',
    'ExpectedReturn', 'ProbabilityOfDefault', 'Rating','Restructured']
    
    interest_df = df[interest_features]
    X = pd.DataFrame(interest_scaler_x.fit_transform(interest_df), columns=interest_df.columns)
    X = X.values.reshape(1, -1)
    
    # Perform the prediction
    interest_model =  joblib.load('C:/Users/KRISHNA/Desktop/SHWETA/Fall 2024/Project/Models/interest_model.joblib')
    y_pred = interest_model.predict(X).reshape(-1, 1)
    print(y_pred)
    
    interest = interest_scaler_y.inverse_transform(y_pred)
    
    #interest = np.expm1(y_pred) # Use this value to show on UI. for Interest amount prediction.
    
    print("Interest Amount: "+ str(interest))
    
    return interest
```
